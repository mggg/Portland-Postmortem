### Add condensed and viable-condensed rankings to precinct shapefile for D1.

### Candidates:

```python
D1_candidates = [
    'Candace Avalos',
    'Jamie Dunphy',
    'Loretta Smith',
    'Noah Ernst',
    'Terrence Hayes',
    'Steph Routh',
    'Timur Ender',
    'Doug Clove',
    'Peggy Sue Owens',
    'David Linn',
    'Joe Allen',
    'Michael (Mike) Sands',
    'Deian Salazar',
    'Cayle Tern',
    'Thomas Shervey',
    'Joe Furi',
    'Uncertified Write In'
]
```

### Viable candidates:

```python
D1_viable_candidates = [
    'Candace Avalos',
    'Jamie Dunphy',
    'Loretta Smith',
    'Noah Ernst',
    'Terrence Hayes',
    'Steph Routh',
    'Timur Ender'
]
```

### Columns for rankings for candidates, plus columns for ballot length:

```python
condensed_score_cols = []
viable_condensed_score_cols = []

for i in range(1, 7):
    condensed_score_cols += [name + " Condensed Rank " + str(i) for name in D1_candidates]
    viable_condensed_score_cols += [name + " Viable-Condensed Rank " + str(i) for name in D1_viable_candidates]

condensed_borda_654321_score_cols = [name + " Condensed Borda 654321" for name in D1_candidates]
viable_condensed_borda_654321_score_cols = [name + " Viable-Condensed Borda 654321" for name in D1_viable_candidates]

condensed_borda_521000_score_cols = [name + " Condensed Borda 521000" for name in D1_candidates]
viable_condensed_borda_521000_score_cols = [name + " Viable-Condensed Borda 521000" for name in D1_viable_candidates]

ballot_length_cols = ["Ballot length 1", "Ballot length 2", "Ballot length 3", "Ballot length 4", "Ballot length 5", "Ballot length 6"]
```

```python
viable_condensed_score_cols
```

```python
viable_condensed_borda_654321_score_cols
```

```python
for col in condensed_score_cols:
    D1_precincts_df.insert(len(D1_precincts_df.columns), col, 0)

D1_precincts_df.insert(len(D1_precincts_df.columns), "Condensed total votes", 0)

for col in condensed_borda_654321_score_cols:
    D1_precincts_df.insert(len(D1_precincts_df.columns), col, 0)

D1_precincts_df.insert(len(D1_precincts_df.columns), "Condensed Borda 654321 total", 0)


for col in condensed_borda_521000_score_cols:
    D1_precincts_df.insert(len(D1_precincts_df.columns), col, 0)

D1_precincts_df.insert(len(D1_precincts_df.columns), "Condensed Borda 521000 total", 0)




for col in viable_condensed_score_cols:
    D1_precincts_df.insert(len(D1_precincts_df.columns), col, 0)

D1_precincts_df.insert(len(D1_precincts_df.columns), "Viable-Condensed total votes", 0)

for col in viable_condensed_borda_654321_score_cols:
    D1_precincts_df.insert(len(D1_precincts_df.columns), col, 0)

D1_precincts_df.insert(len(D1_precincts_df.columns), "Viable-Condensed Borda 654321 total", 0)


for col in viable_condensed_borda_521000_score_cols:
    D1_precincts_df.insert(len(D1_precincts_df.columns), col, 0)

D1_precincts_df.insert(len(D1_precincts_df.columns), "Viable-Condensed Borda 521000 total", 0)





for col in ballot_length_cols:
    D1_precincts_df.insert(len(D1_precincts_df.columns), col, 0)

D1_precincts_df.insert(len(D1_precincts_df.columns), "Ballot length total", 0)

```

```python
D1_precincts_df
```

### Add scores!

```python
for p in tqdm(D1_precincts_df.index):
    this_precinct = D1_precincts_df.loc[p, 'Precinct']
    this_precinct_condensed_cvr = D1_condensed_cvr[D1_condensed_cvr['Precinct'] == this_precinct]
    this_precinct_viable_condensed_cvr = D1_viable_condensed_cvr[D1_viable_condensed_cvr['Precinct'] == this_precinct]
    
    for b in this_precinct_condensed_cvr.index:
        ballot_r1 = this_precinct_condensed_cvr.loc[b, "rank1"]
        ballot_r2 = this_precinct_condensed_cvr.loc[b, "rank2"]
        ballot_r3 = this_precinct_condensed_cvr.loc[b, "rank3"]
        ballot_r4 = this_precinct_condensed_cvr.loc[b, "rank4"]
        ballot_r5 = this_precinct_condensed_cvr.loc[b, "rank5"]
        ballot_r6 = this_precinct_condensed_cvr.loc[b, "rank6"]

        # Condensed rank scores:
        if ballot_r1 in D1_candidates:
            D1_precincts_df.loc[p, ballot_r1 + " Condensed Rank 1"] += 1
            D1_precincts_df.loc[p, "Condensed total votes"] += 1 
        if ballot_r2 in D1_candidates:
            D1_precincts_df.loc[p, ballot_r2 + " Condensed Rank 2"] += 1
        if ballot_r3 in D1_candidates:
            D1_precincts_df.loc[p, ballot_r3 + " Condensed Rank 3"] += 1
        if ballot_r4 in D1_candidates:
            D1_precincts_df.loc[p, ballot_r4 + " Condensed Rank 4"] += 1
        if ballot_r5 in D1_candidates:
            D1_precincts_df.loc[p, ballot_r5 + " Condensed Rank 5"] += 1
        if ballot_r6 in D1_candidates:
            D1_precincts_df.loc[p, ballot_r6 + " Condensed Rank 6"] += 1

        # Ballot length scores:
        num_skips = [ballot_r1, ballot_r2, ballot_r3, ballot_r4, ballot_r5, ballot_r6].count("skipped")
        ballot_length = 6 - num_skips
        D1_precincts_df.loc[p, "Ballot length " + str(ballot_length)] += 1
        D1_precincts_df.loc[p, "Ballot length total"] += 1


    for b in this_precinct_viable_condensed_cvr.index:
        ballot_r1 = this_precinct_viable_condensed_cvr.loc[b, "rank1"]
        ballot_r2 = this_precinct_viable_condensed_cvr.loc[b, "rank2"]
        ballot_r3 = this_precinct_viable_condensed_cvr.loc[b, "rank3"]
        ballot_r4 = this_precinct_viable_condensed_cvr.loc[b, "rank4"]
        ballot_r5 = this_precinct_viable_condensed_cvr.loc[b, "rank5"]
        ballot_r6 = this_precinct_viable_condensed_cvr.loc[b, "rank6"]

        # Viable-Condensed rank scores:
        if ballot_r1 in D1_viable_candidates:
            D1_precincts_df.loc[p, ballot_r1 + " Viable-Condensed Rank 1"] += 1
            D1_precincts_df.loc[p, "Viable-Condensed total votes"] += 1 
        if ballot_r2 in D1_viable_candidates:
            D1_precincts_df.loc[p, ballot_r2 + " Viable-Condensed Rank 2"] += 1
        if ballot_r3 in D1_viable_candidates:
            D1_precincts_df.loc[p, ballot_r3 + " Viable-Condensed Rank 3"] += 1
        if ballot_r4 in D1_viable_candidates:
            D1_precincts_df.loc[p, ballot_r4 + " Viable-Condensed Rank 4"] += 1
        if ballot_r5 in D1_viable_candidates:
            D1_precincts_df.loc[p, ballot_r5 + " Viable-Condensed Rank 5"] += 1
        if ballot_r6 in D1_viable_candidates:
            D1_precincts_df.loc[p, ballot_r6 + " Viable-Condensed Rank 6"] += 1
            

# Borda scores and totals:

for name in D1_candidates:
    D1_precincts_df[name + " Condensed Borda 654321"] = 6*D1_precincts_df[name + " Condensed Rank 1"] + 5*D1_precincts_df[name + " Condensed Rank 2"] + 4*D1_precincts_df[name + " Condensed Rank 3"] + 3*D1_precincts_df[name + " Condensed Rank 4"] + 2*D1_precincts_df[name + " Condensed Rank 5"] + 1*D1_precincts_df[name + " Condensed Rank 6"]
    D1_precincts_df["Condensed Borda 654321 total"] = D1_precincts_df["Condensed Borda 654321 total"] + D1_precincts_df[name + " Condensed Borda 654321"]

    D1_precincts_df[name + " Condensed Borda 521000"] = 5*D1_precincts_df[name + " Condensed Rank 1"] + 2*D1_precincts_df[name + " Condensed Rank 2"] + 1*D1_precincts_df[name + " Condensed Rank 3"] 
    D1_precincts_df["Condensed Borda 521000 total"] = D1_precincts_df["Condensed Borda 521000 total"] + D1_precincts_df[name + " Condensed Borda 521000"]

for name in D1_viable_candidates:
    D1_precincts_df[name + " Viable-Condensed Borda 654321"] = 6*D1_precincts_df[name + " Viable-Condensed Rank 1"] + 5*D1_precincts_df[name + " Viable-Condensed Rank 2"] + 4*D1_precincts_df[name + " Viable-Condensed Rank 3"] + 3*D1_precincts_df[name + " Viable-Condensed Rank 4"] + 2*D1_precincts_df[name + " Viable-Condensed Rank 5"] + 1*D1_precincts_df[name + " Viable-Condensed Rank 6"]
    D1_precincts_df["Viable-Condensed Borda 654321 total"] = D1_precincts_df["Viable-Condensed Borda 654321 total"] + D1_precincts_df[name + " Viable-Condensed Borda 654321"]

    D1_precincts_df[name + " Viable-Condensed Borda 521000"] = 5*D1_precincts_df[name + " Viable-Condensed Rank 1"] + 2*D1_precincts_df[name + " Viable-Condensed Rank 2"] + 1*D1_precincts_df[name + " Viable-Condensed Rank 3"] 
    D1_precincts_df["Viable-Condensed Borda 521000 total"] = D1_precincts_df["Viable-Condensed Borda 521000 total"] + D1_precincts_df[name + " Viable-Condensed Borda 521000"]

```
